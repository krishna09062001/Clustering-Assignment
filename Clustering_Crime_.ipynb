# **Hierarchical Clustering**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
crime=pd.read_csv('crime_data.csv')
crime
```

```python
crime.info()
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
crime
```

```python
# Normalize heterogenous numerical data
crime_norm=pd.DataFrame(normalize(crime),columns=crime.columns)
crime_norm
```

```python
# Create Dendrograms
plt.figure(figsize=(10, 7))
dendograms=sch.dendrogram(sch.linkage(crime_norm,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
hclusters
```

```python
import warnings
warnings.filterwarnings('ignore')
y=pd.DataFrame(hclusters.fit_predict(crime_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
crime['clustersid']=hclusters.labels_
crime
```

```python
crime.groupby('clustersid').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))
plt.scatter(crime['clustersid'],crime['Murder'], c=hclusters.labels_)
```

# **K Means Clustering**

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
crime.head()
```

```python
crime.info()
```

```python
# Normalize heterogenous numerical data by using Standard Scaler
crime_norm=StandardScaler().fit_transform(crime)
```

```python
# within-cluster sum-of-squares criterion
import warnings
warnings.filterwarnings('ignore')
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(crime_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Cluster algorithm using K=3
clusters3=KMeans(3,random_state=30).fit(crime_norm)
clusters3
```

```python
clusters3.labels_
```

```python
# Assign clusters to the data set
crime1=crime.copy()
crime1['clusters3id']=clusters3.labels_
crime1
```

```python
# Compute the centroids for K=4 clusters with 11 variables
clusters3.cluster_centers_
```

```python
# Group data by Clusters (K=4)
crime1.groupby('clusters3id').agg(['mean']).reset_index()
```

```python
#Plot Clusters
plt.figure(figsize=(10, 7))
plt.scatter(crime1['clusters3id'],crime1['Murder'], c=clusters3.labels_)
```

```python
# Cluster algorithm using K=5
clusters5=KMeans(5,random_state=30).fit(crime_norm)
clusters5
```

```python
clusters5.labels_
```

```python
crime2=crime.copy()
crime2['clusters5id']=clusters5.labels_
crime2
```

```python
clusters5.cluster_centers_
```

```python
crime2.groupby('clusters5id').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(crime2['clusters5id'],crime2['Murder'], c=clusters5.labels_)
```

# **DBSCAN Clustering**

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime.head()
```

```python
crime.info()
```

```python
# Normalize heterogenous numerical data using standard scalar fit transform to dataset
crime_norm=StandardScaler().fit_transform(crime)
crime_norm
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(crime_norm)
```

```python
dbscan.labels_
```

```python
# Adding clusters to dataset
crime['clusters']=dbscan.labels_
crime
```

```python
crime.groupby('clusters').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))
plt.scatter(crime['clusters'],crime['UrbanPop'], c=dbscan.labels_)
```
