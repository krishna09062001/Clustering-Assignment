# **Hierarchical Clustering**

# Using Normalize function

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
Airlines=pd.read_csv('EastWestAirlines.csv')
Airlines
```

```python
Airlines.info()
```

```python
Airline=Airlines.drop(['ID'],axis=1)
Airline
```

```python
# Normalize heterogenous numerical data
Airline_norm=pd.DataFrame(normalize(Airline),columns=Airline.columns)
Airline_norm
```

```python
# Create Dendrograms
plt.figure(figsize=(10, 7))
dendograms=sch.dendrogram(sch.linkage(Airline_norm,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
hclusters
```

```python
AgglomerativeClustering(n_clusters=5)
```

```python
import warnings
warnings.filterwarnings('ignore')
y=pd.DataFrame(hclusters.fit_predict(Airline_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
Airline['clustersid']=hclusters.labels_
Airline
```

```python
Airline.groupby('clustersid').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))
plt.scatter(Airline['clustersid'],Airline['Balance'], c=hclusters.labels_)
```

# **K Means Clustering**

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
airline=Airlines.drop(['ID'],axis=1)
airline
```

```python
airline.info()
```

```python
# Normalize heterogenous numerical data by using Standard Scaler
airline2_norm=StandardScaler().fit_transform(airline)
```

```python
# within-cluster sum-of-squares criterion
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(airline2_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# **Build Cluster Algorithm using K=4**

```python
# Cluster algorithm using K=4
clusters4=KMeans(4,random_state=30).fit(airline2_norm)
clusters4
```

```python
clusters4.labels_
```

```python
# Assign clusters to the data set
airline4=airline.copy()
airline4['clusters4id']=clusters4.labels_
airline4
```

```python
# Compute the centroids for K=4 clusters with 11 variables
clusters4.cluster_centers_
```

```python
# Group data by Clusters (K=4)
airline4.groupby('clusters4id').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))
plt.scatter(airline4['clusters4id'],airline4['Balance'], c=clusters4.labels_)
```

# **DBSCAN Clustering**

```python
# Import Libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
Airlines.drop(['ID'],axis=1,inplace=True)
Airlines
```

```python
# Normalize heterogenous numerical data using standard scalar fit transform to dataset
airline_norm=StandardScaler().fit_transform(airline)
airline_norm
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(airline_norm)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
# Adding clusters to dataset
airline['clusters']=dbscan.labels_
airline
```

```python
airline.groupby('clusters').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))
plt.scatter(airline['clusters'],airline['Bonus_miles'], c=dbscan.labels_)
```
